## Removendo linhas desnecessÃ¡rias, e alterando nome das colunas:

```python
fechamentos_limpo = fechamentos

fechamentos_limpo.columns = fechamentos_limpo.iloc[2]

fechamentos_limpo = fechamentos_limpo.drop(fechamentos_limpo.index[:3])

fechamentos_limpo.head(20)
```

```python

fechamentos_limpo.columns = fechamentos_limpo.columns.str.replace('Fechamento\najust p/ prov\nEm moeda orig\n','', regex=True) #"Apagando" informaÃ§Ãµes irrelevantes do nome da coluna, deixando apenas o nome da aÃ§Ã£o
fechamentos_limpo.head()
```

## Corrigindo o indice:

```python
fechamentos_limpo['Data'] = pd.to_datetime(fechamentos_limpo['Data'])

fechamentos_limpo = fechamentos_limpo.set_index('Data')

fechamentos_limpo.head(20)
```
